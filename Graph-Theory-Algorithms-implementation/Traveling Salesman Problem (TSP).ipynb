### ACM : Prim-Jarnik

```python
from math import inf 
import numpy as np
```

```python
def PrimJarnik(G,s):
    T=[]
    n=len(G)
    Q={0:[s,(0,0)]}
    for i in range(n):
        if i!=s:
            Q[np.sum(G)+i]=[i,(0,0)]
    L=Q.keys()
    while Q!={}:
        m=min(L)
        q=Q[m][0]
        t=Q[m]
        del Q[m]
        T+=[t]
        A=G[t[0]]
        for e in A:
            if e!=0:
                j=A.index(e)
                for e in Q.items():
                    e=list(e)
                    if e[1][0]==j:
                        if A[j]<e[0]:
                            e[0]=A[j]
                            e[1][1]=(q,j)
    return T  
```

```python
G=[[0,2,8,0,0,7],[2,0,5,7,0,0],[8,5,0,9,0,8],[0,7,7,0,4,0],[0,0,0,4,0,3],[7,0,8,0,3,0]]
n=len(G)
Q={0:[0,(0,0)]}
for i in range(n):
    if i!=0:
        Q[np.sum(G)+i]=[i,(0,0)]
Q.items()
```

### TSP : Voyageur de commerce 

```python
A=[[0,5,6,8],[0,5,6,8],[5,0,10,4],[8,4,7,0]] # Condition de l'inégalité triangulaire vérifiée 
PrimJarnik(A,0)
```

```python
def TSP(G,s): # ce qu'apporte l'algo de Prim à cette approche c'est l'odre de passage par les points pour construire l'ACM 
    L=PrimJarnik(G,s)
    H=[]
    K=[]
    for e in L:
        q=e[1]
        H+=[q]
    for e in H:
        if e[0] not in K:
            K+=[e[0]]
        if e[1] not in K:
            K+=[e[1]]
    K.append(s)
    return K      
```

```python
TSP(A,0)
```

```python
def ParcoursProfondeur(R,s):
    n=len(R)
    visites=[False] * n
    visites[s]=True
    print(s, end=' ,')
    for i in range(len(R[s])):
        if R[s][i]!=0 and visites[i] == False:
            ParcoursProfondeur(R,i)
```
